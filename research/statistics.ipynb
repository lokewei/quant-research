```python
import numpy as np
import matplotlib.pyplot as plt

data_unionform = np.random.random_sample((1000))
plt.hist(data_unionform, bins=30)
plt.show()
```

```python
data_normal = np.random.standard_normal((10000))
plt.hist(data_normal, bins=30)
plt.show()
```
